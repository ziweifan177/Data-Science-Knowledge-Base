```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import seaborn as sns

dataset=pd.DataFrame(
    pd.read_csv(r"C:\Users\Zi Wei Fan\Desktop\NEU\NEU\Course\CSYE7245_BigData&IntelligenceAnalytics\Assignment\7245_Assignment1\weather-data\weather_data_nyc_centralpark_2016.csv"))
print(dataset.head(10))

#Check if any NaN?
print(np.any(dataset.isnull==True))
print("Numbers of NaN:")
print(dataset.isnull().sum())

#Fill NaN:
dataset=dataset.fillna(value=0)
print(dataset.head(10))
dataset=dataset[["maximum_temperature","precipiation"]]

#Detect outlier with DBSCAN:
dbscan=DBSCAN(eps=1,min_samples=15).fit(dataset)
labels=dbscan.labels_
# labels_ : array, shape = [n_samples]
# Cluster labels for each point in the dataset given to fit(). 
# Noisy samples are given the label -1.
coreSamples=np.zeros_like(labels,dtype=bool) #FALSE matrix.
coreSamples[dbscan.core_sample_indices_]=True
labels=dbscan.labels_

#Number of clusters in labels, ignoring noise if present.
n_clusters_=len(set(labels))-(1 if -1 in labels else 0)
#the set() function can be used to create sets. 

#Plot:
sns.set(style="ticks",color_codes=True)
unique_labels=set(labels)
colors=[plt.cm.Spectral(each)
        for each in np.linspace(0, 1, len(unique_labels))]
#matplotlib.cm:
# This module provides a large set of colormaps, functions for registering new colormaps and for getting a colormap by name, and a mixin class for adding color mapping functionality.
#Spectral():
#set the default colormap to spectral and apply to current image if any. 
#np.linspace():
#Return evenly spaced numbers over a specified interval.

print('Scatter Plot:')
ax1=dataset.plot.scatter(x='maximum_temperature',y='precipiation',label='Precipiation-Temperature Chart')
plt.show()
```

## 3. Plot each colmun as appropriate for the data type:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import seaborn as sns

plt.rcParams['figure.figsize']=(10,10)
dataset=pd.read_csv(r"7245_Assignment1\weather-data\weather_data_nyc_centralpark_2016.csv")
dataset=dataset.fillna(value=0)

print('1. plot.Scatter: Maximum_temperature & precipiation:')
data=dataset[["maximum_temperature","precipiation"]]
ax1=data.plot.scatter(x='maximum_temperature',y='precipiation',color='DarkBlue',label='MaxTemperature_precipiation Plot')
plt.show()

print('2. sns.Pairplot: minimum_temperature & precipiation:')
data=dataset[['minimum_temperature','precipiation']]
ax2=sns.pairplot(data)
plt.show()

print('3. plot.Scatter: avg_temperature & precipiation:')
data=dataset[['average_temperature','precipiation']]
ax3=data.plot.scatter(x='average_temperature',y='precipiation',color='Purple',label='Avg temperature & precipiation Chart')
plt.show()

print('4. Joint plot: precipiation & Snow fall:')
data=dataset[['precipiation','snow_fall']]
sns.set(style="white", color_codes=True)
#tips = sns.load_dataset("tips")
g = sns.jointplot(x="precipiation", y="snow_fall", data=data)
plt.show()

print('5. Box plot: average_temperature & Snow fall:')
data=dataset[['average_temperature','snow_fall']]
ax5=sns.boxplot(x='average_temperature',y='snow_fall',data=data)
plt.show()

print('6. Joint Plot with Reg: Relationship between Max_Temperature & Avg Temperature:')
data=dataset[['average_temperature','maximum_temperature']]
ax6=sns.jointplot(x='average_temperature',y='maximum_temperature',data=data,kind='reg')
plt.show()

print('7. Joint plot with hex: Min_Temperature & Avg Temperature:')
data=dataset[['average_temperature','minimum_temperature']]
ax6=sns.jointplot(x='average_temperature',y='minimum_temperature',data=data,kind='hex')
plt.show()

print('8. Joint plot with reg: Date & Preicipiation:')
data=dataset[['date','precipiation']]
ax7=data.set_index('date').plot()
plt.show()
```
